## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

_______________________________________________________________________________________________________________________________
A primary key is a unique identifier for a table that has to be comprised of unique values for that column. The primary key value is stored in an index and can neither be blank or null. A foriegn key is a key that refers to a  primary key that appears on a another table. 




## Question 3

Explain the different types of relationships entities can have in a SQL database.

_______________________________________________________________________________________________________________________________
The different types of relationships are one to one in which one record in the first table corresponds to one record in the second table and vice versa, one to many where one record on the first table relates to many records on the second table and records on the second table table only relate to one record on the first table, and lastly many to many where one record on the first table relates to many in the second table and one in the second table relates to many in the first table.


## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Inner join, selects all records from two tables in which the join condition is met. Left join selects all records from left table and all records from both tables in which join condition is met. Right join selects all records from right table and all records from both tables in which join condition is met. Full join selects all records from both tables whether join conditions are met or not.
